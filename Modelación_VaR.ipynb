Estas entidades, por su propia naturaleza operativa, constituyen verdaderas fábricas de riesgo. Cada producto, servicio o decisión financiera conlleva una exposición inherente que debe ser identificada, cuantificada, monitoreada y gestionada con rigurosidad. Esta secuencia constituye el núcleo del proceso de gestión integral de riesgos, cuyo objetivo último es proteger la rentabilidad y la solvencia patrimonial de la organización.

El riesgo, en esencia, debe entenderse como la probabilidad de que eventos adversos, previstos o imprevistos, impacten negativamente en los objetivos estratégicos de la entidad. Gestionar el riesgo implica aceptar que cualquier decisión –sea una inversión, una operación crediticia, una política de fondeo o una estrategia operativa– genera una distribución de rendimientos potenciales. Como analistas y gestores, nuestro deber es conocer y cuantificar esa distribución, estimar su rendimiento esperado, y calcular la posible desviación respecto a dicho valor medio.

No obstante, desde una perspectiva de gestión prudencial, nuestro foco debe centrarse en el lado negativo de esa variabilidad, es decir, en la probabilidad de obtener resultados inferiores a lo esperado. Es aquí donde cobra especial relevancia una de las herramientas más robustas y ampliamente utilizadas en la gestión del riesgo financiero: el Valor en Riesgo (VaR).

El VaR es una medida estadística diseñada para responder una pregunta fundamental en la gestión del riesgo:

**¿Cuál es la peor pérdida que podríamos esperar, bajo condiciones normales de mercado, durante un horizonte temporal específico y con un nivel de confianza determinado?**

Esta métrica proporciona una estimación de la pérdida máxima esperada en condiciones normales, y se ha consolidado como una herramienta clave para evaluar distintos tipos de riesgos en una institución financiera:

Riesgo de mercado, al estimar las posibles pérdidas derivadas de movimientos

*   Riesgo de mercado, al estimar las posibles pérdidas derivadas de movimientos adversos en tasas de interés, precios de activos o tipos de cambio.
*   Riesgo operativo, cuando se aplica a la medición de pérdidas recurrentes asociadas a procesos, personas, sistemas o eventos externos.
*   Riesgo de liquidez, en la evaluación de salidas máximas de fondos (por ejemplo, retiros de depósitos) en situaciones de estrés.

En síntesis, el VaR representa una forma disciplinada y cuantitativa de anticiparse al impacto financiero de escenarios adversos, lo que permite a la Alta Gerencia y a los responsables de riesgo tomar decisiones informadas, alineadas con el perfil de riesgo y la estrategia institucional.

Existen dos tipos de métodos para calcular el VAR:

1.   Método No Paramétricos
2.   Métodos Paramétricos

**Métodos Paramétricos**

Los métodos paramétricos para estimar el VaR se basan en suponer que los rendimientos de los activos siguen una distribución de probabilidad específica, comúnmente la normal. Estos métodos requieren estimar parámetros como la media y la desviación estándar de los rendimientos.

**Métodos No Paramétricos**

Los métodos no paramétricos no hacen suposiciones específicas sobre la distribución de los rendimientos. Uno de los enfoques más comunes es la simulación histórica, que utiliza datos reales de rendimientos pasados para estimar el VaR. Este método implica ordenar los rendimientos históricos y determinar el percentil correspondiente al nivel de confianza seleccionado.​

La estimación del Valor en Riesgo (VaR) ha evolucionado considerablemente en las últimas décadas, dando lugar a una diversidad de métodos que permiten adaptarse a las características de cada institución, a la naturaleza del riesgo evaluado y a la disponibilidad de datos históricos. No existe un enfoque único o universalmente superior; por el contrario, la elección del método más apropiado debe considerar tanto la complejidad del portafolio como el nivel de precisión deseado en la captura de eventos extremos.

Entre los métodos más utilizados en la práctica profesional se encuentran los siguientes:

**Método Histórico:**

Este enfoque parte del supuesto de que el comportamiento pasado del mercado es representativo del futuro. Se basa en los rendimientos históricos observados y no asume una distribución estadística específica. Su principal ventaja radica en su simplicidad y en que captura eventos extremos tal como ocurrieron. No obstante, su precisión está sujeta a la calidad y extensión de la base de datos utilizada.

**Método de Varianza-Covarianza (o método paramétrico):**

Este modelo, ampliamente difundido por su eficiencia computacional, parte del supuesto de que los rendimientos siguen una distribución normal. Requiere únicamente el cálculo de la media y la desviación estándar de la serie, y en casos de portafolios múltiples, la matriz de varianzas y covarianzas. Si bien es útil en escenarios con baja volatilidad y donde la normalidad es razonable, subestima el riesgo en presencia de colas pesadas o asimetrías.

**Simulación de Monte Carlo:**

Este enfoque utiliza simulaciones aleatorias para modelar miles (o millones) de posibles escenarios futuros del portafolio, basados en supuestos estadísticos definidos por el analista. Su fortaleza radica en la flexibilidad y capacidad de incorporar dinámicas complejas y distribuciones no normales, aunque su implementación requiere mayor capacidad computacional y un sólido conocimiento en modelado estocástico.

**Modelos ARCH, GARCH y extensiones:**

En contextos donde la volatilidad es dinámica y presenta heterocedasticidad condicional, los modelos ARCH (Autoregressive Conditional Heteroskedasticity) y GARCH (Generalized ARCH) se convierten en herramientas superiores. Estos modelos permiten capturar la naturaleza cambiante de la volatilidad en el tiempo, reflejando de forma más realista el riesgo latente en condiciones de mercado adversas. Su aplicación en la estimación del VaR permite una representación más precisa de eventos extremos.

# **3. Pruebas de Distribución**

Se ha realizado la prueba de Kolmogorov-Smirnov. La prueba KS es un test estadístico que se utiliza para comparar una muestra con una distribución teórica específica, evaluando si los datos siguen esa distribución:

*   Hipótesis nula (H0): Los datos provienen de la distribución teórica específicada.
*   Hipótesis alternativa (H1): Los datos no provienen de esa distribución.

El test calcula un estadístico KS que mide la máxima diferencia entre la función de distribución empírica de los datos y la función de distribución teórica. El p-valor asociado indica la probabilidad de observar esa diferencia (o una mayor) si la hipótesis nula fuera cierta.

**Interpretación:**

*   En todos los casos, el p-valor es menor que 0.05 (nivel de significancia comúnmente usado), lo que implica que se rechaza la hipótesis nula para todas las distribuciones probadas.
*   Esto significa que ninguna de las distribuciones ajustadas (gamma, normal, lognormal, weibull_min, exponencial, beta) representa adecuadamente la distribución de tus datos según la prueba KS.
*  El estadístico KS más bajo se observa en las distribuciones gamma, normal y lognormal (aproximadamente 0.115), lo que indica que estas distribuciones se ajustan mejor que las demás, pero aún así el ajuste no es estadísticamente aceptable con base en el p-valor.






```python
#Pruebas de ajuste de distribución (Kolmogorov-Smirnov)
distribuciones = ['norm', 'lognorm', 'expon', 'gamma', 'beta', 'weibull_min']
resultados = []

for dist_name in distribuciones:
    dist = getattr(stats, dist_name)
    params = dist.fit(serie)
    D, p = stats.kstest(serie, dist_name, args=params)
    resultados.append({
        "Distribución": dist_name,
        "Estadístico KS": round(D, 4),
        "p-valor": round(p, 4)
    })

#Convertimos a DataFrame para mostrar en forma de tabla ordenada
resultados_df = pd.DataFrame(resultados)
resultados_df = resultados_df.sort_values(by='p-valor', ascending=False)

print("\nResultados de la prueba de Kolmogorov-Smirnov:\n")
print(resultados_df)

#Gráfico de distribución de los datos
plt.figure(figsize=(10,6))
sns.histplot(serie, kde=True, stat="density", bins=30, color='Mediumblue')
plt.title('Distribución de los Datos')
plt.xlabel('Valores')
plt.ylabel('Densidad o Frecuencia')
plt.grid(True)
plt.show()
```

# **4. Cáculo de Variaciones Logaritmicas**

Como se ha podido mencionar lineas arriba para la aplicación del VaR por el metodo de Varianza y Covarianza se asume que la distribución de los datos es normal por tal motivo es requerido calcular las variaciones logaritmacas de periodo a periodo.

**Propiedades estadísticas favorables:**

Los rendimientos logarítmicos tienden a estar más cerca de una distribución normal.
La suma de rendimientos logarítmicos en varios periodos equivale al rendimiento logarítmico total del periodo largo, lo que facilita el análisis de riesgos en diferentes horizontes temporales.

Se utilizan las variaciones logarítmicas para calcular el VaR paramétrico porque:

1.   Permiten suponer normalidad (o aproximación) en los rendimientos.
1.   Hacen el análisis matemático y estadístico más robusto y coherente.
2.   Son mas apropiada para el manejo de riesgo y comparaciones en series financieras.

Tal como indica la teoría al aplicar la logaritmización se ha logrado obtener una distribución proxima a la normalidad, tal como se podra obser en el Gráfico Q - Q, donde:

*   Los puntos azules se alinean mucho mejor a la línea roja, especialmente en la parte central del gráfico.
*   Hay ligeras desviaciones en los extremos (colas), pero la mayor parte de los datos sigue la línea de referencia.
*    Esto indica que los rendimientos logarítmicos se aproximan bastante bien a una distribución normal, aunque hay algunas excepciones en los valores más extremos (lo cual es común en series financieras).

Por tanto transformar la serie original a rendimientos logarítmicos mejora notablemente el ajuste a la normalidad.

Esto valida el uso de modelos paramétricos (como el VaR paramétrico) que asumen normalidad en los rendimientos.

Sin embargo, es importante tener en cuenta que los extremos (colas) presentan ligeras desviaciones, lo que sugiere que, aunque la normalidad es una buena aproximación, podrían existir riesgos extremos no capturados completamente por el modelo normal por lo que se podría subestimar la exposición al riesgo de el volumen de salida de depósitos a un determinado nivel de confianza para el caso de estudio particularmente, pudiendo inducir a una toma de decisiones incorrecta en la gestión de riesgo de liquidez y mercado principalmente.





    

   







```python
# Calculamos los log-retornos
df['Log_Retornos'] = np.log(df['Depositos'] / df['Depositos'].shift(1))

# Eliminamos valores NaN
df = df.dropna()

# Graficar los log-retornos
plt.figure(figsize=(10, 4))
plt.plot(df['Log_Retornos'], label='Log-Retornos', color='Mediumblue')
plt.title('Serie de Log-Retornos de Depósitos')
plt.xlabel('Tiempo')
plt.ylabel('Log-Retornos')
plt.legend()
plt.grid(True)
plt.show()

# Gráfico de Distrubición de los Retorno Logaritmicos en Depósitos

plt.figure(figsize=(8, 4))
sns.histplot(df['Log_Retornos'], kde=True, bins=20, color='Mediumblue')
plt.title('Distribución de los Log-Retornos')
plt.xlabel('Log-Retornos')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Prueba de normalidad: Shapiro-Wilk y test de Kolmogorov-Smirnov
shapiro_test = stats.shapiro(df['Log_Retornos'])
ks_test = stats.kstest(df['Log_Retornos'], 'norm', args=(df['Log_Retornos'].mean(), df['Log_Retornos'].std()))

print("Shapiro-Wilk Test:")
print(f"  Estadístico: {shapiro_test.statistic:.4f}, p-valor: {shapiro_test.pvalue:.4f}")

print("\nKolmogorov-Smirnov Test:")
print(f"  Estadístico: {ks_test.statistic:.4f}, p-valor: {ks_test.pvalue:.4f}")

#Gráfico Q-Q para verificar si hay ajuste a la normal
sm.qqplot(df['Log_Retornos'], line='s', color='Mediumblue')
plt.title('Gráfico Q-Q de la Serie Logaritmica')
plt.grid(True)
plt.show()
```

# **5. Cáculo de Variaciones Logaritmicas**

El Valor en Riesgo (VaR) es una herramienta estadística fundamental en la gestión de riesgos financieros, que permite cuantificar la máxima pérdida esperada de una posición o portafolio bajo condiciones normales de mercado, para un horizonte temporal y un nivel de confianza determinados.

En el contexto del presente estudio, el VaR representa la máxima salida o retiro de depósitos que podría experimentar el Sistema Financiero Boliviano, con un nivel de confianza del 95%. Esto significa que, bajo las condiciones históricas observadas y asumiendo la validez del modelo, los retiros de depósitos alcanzarian los 6.176.183.695,84 Bob en el periodo analizado con un 95% de confianza. Alternativamente, hay un 5% de probabilidad de que los retiros excedan dicho valor, lo que constituye el riesgo extremo que el sistema podría enfrentar.
Fórmula empleada

La estimación del VaR paramétrico se realizó utilizando la siguiente expresión:

         VaR=Último Valor de Depósitos*Z*Desv_Est Variaciones Logaritmicas


```python
mu = df['Log_Retornos'].mean() #Media
std_Depositos=df['Log_Retornos'].std() #Desviacion
valor_z=norm.ppf(1-0.95) #Valor z
ultimo_valor=df['Depositos'].iloc[-1] #Ultimo Valor

#Formula de VaR Parametrico
VaR_Parametrico=valor_z*std_Depositos*ultimo_valor #Var Parametrico
print(f"Media = {mu:.4f}")
print(f"Desviación Estandar = {std_Depositos:.4f}")
print(f"Ultimo Deposito = {ultimo_valor:,.2f}")
print(f"Valor en Riesgo Parametrico = {VaR_Parametrico:,.2f}")
```

Como análisis complementario, se identificaron los periodos en los que las salidas de depósitos superaron el Valor en Riesgo (VaR) estimado, encontrándose que únicamente en el periodo de abril de 2023 se registró una pérdida superior al umbral calculado. Este evento coincide con la intervención del Ex Banco Fasil, lo que evidencia que dicho acontecimiento representó un episodio de riesgo extremo para el sistema financiero boliviano, estadísticamente poco frecuente según la metodología VaR empleada.

```python
# Calcular las pérdidas reales (log-retornos)
df['Pérdidas_Reales'] = df['Log_Retornos'] * df['Depositos']

# Filtrar pérdidas mayores que el VaR
df['Pérdidas_Mayores_que_VaR'] = df['Pérdidas_Reales'] < VaR_Parametrico

# Gráfico comparativo de log-retornos vs. VaR
plt.figure(figsize=(10, 6))
sns.histplot(df['Log_Retornos'], kde=True, bins=30, color='Mediumblue', label='Log-Retornos')
plt.axvline(VaR_Parametrico/ultimo_valor, color='red', linestyle='--', label=f'VaR Paramétrico ({VaR_Parametrico:,.2f})')
plt.title('Comparación de Log-Retornos con VaR Paramétrico')
plt.xlabel('Log-Retornos')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()

# Mostrar las filas donde las pérdidas fueron mayores que el VaR
print("\nPérdidas mayores que el VaR Paramétrico:")
print(df[df['Pérdidas_Mayores_que_VaR']])
```

El objetivo principal de este enfoque es anticipar posibles pérdidas que podrían superar la tolerancia al riesgo de la entidad, permitiendo así desarrollar planes de contingencia efectivos y fortalecer la toma de decisiones estratégicas por parte de la Alta Gerencia.

## **Método de Monte Carlo**
El método de Monte Carlo es una técnica de simulación estadística que permite estimar el VaR generando un gran número de escenarios posibles para la evolución futura de los precios o rendimientos de un activo, cartera o variable financiera. A diferencia de los métodos paramétricos (que suponen una distribución específica, como la normal), Monte Carlo puede adaptarse a cualquier distribución y dinámica de precios, capturando mejor la complejidad y los riesgos extremos.

Se asume que los rendimientos futuros siguen una cierta dinámica estocástica. El modelo más simple es el Movimiento Browniano Geométrico:

imagen.png

El método de Monte Carlo para VaR es una herramienta estadística avanzada que permite estimar el riesgo de pérdida máxima bajo escenarios de mercado simulados, sin depender de supuestos rígidos sobre la distribución de los rendimientos. Su fortaleza radica en la flexibilidad y en la capacidad de capturar riesgos extremos y dinámicas complejas, siendo especialmente útil cuando los datos empíricos muestran desviaciones de la normalidad, como se ha observado en el gráfico Q-Q.

**Resultados Obtenidos:**

Los valores obtenidos representan la estimación de la máxima salida de depósitos esperada en el valor total de los depósitos del sistema financiero boliviano bajo diferentes niveles de confianza, utilizando la simulación de Monte Carlo. Esta técnica genera una gran cantidad de escenarios posibles para los cambios futuros en los depósitos, basándose en la variabilidad histórica observada, y calcula la pérdida máxima que no se espera superar con cierta probabilidad.



1.  VaR al 95% (-8,497 millones de bolivianos): Esto significa que, con un 95% de confianza, la máxima salida de depósitos que podría ocurrir en un mes no superará aproximadamente los 8.492.096.256 Bob. En otras palabras, hay un 5% de probabilidad de que la pérdida sea mayor a esta cifra en un mes dado.
2.      VaR al 98% (-10,090 millones de bolivianos) y al 99% (-11,171 millones de bolivianos): A medida que aumenta el nivel de confianza, el VaR también aumenta en magnitud, reflejando escenarios más extremos pero menos probables. Por ejemplo, con un 99% de confianza, la máxima salida de depósitos esperada podría llegar a ser superior a -11.141.741.100 Bob, aunque con solo un 1% de probabilidad de ocurrencia en un mes.
3.   Media de pérdidas simuladas (-2,204 millones de bolivianos): Este valor indica la pérdida promedio esperada en todos los escenarios simulados, incluyendo tanto pérdidas pequeñas como grandes. Es considerablemente menor que los VaR, ya que estos últimos se enfocan en las pérdidas extremas.






    



```python
# Parámetros base
mu = df['Log_Retornos'].mean()
sigma = df['Log_Retornos'].std()
ultimo_valor = df['Depositos'].iloc[-1]
n_simulaciones = 100000  # Número de simulaciones

# Simulación de log-retornos
simulaciones = np.random.normal(loc=mu, scale=sigma, size=n_simulaciones)

# Convertir log-retornos en valores simulados de depósitos
valores_simulados = ultimo_valor * np.exp(simulaciones)

# Calcular pérdidas simuladas
perdidas = ultimo_valor - valores_simulados

# Calcular VaR para diferentes niveles de confianza
VaR_95 = np.percentile(perdidas, 5)
VaR_98 = np.percentile(perdidas, 2)
VaR_99 = np.percentile(perdidas, 1)

# Calcular la media de las pérdidas simuladas
media_perdida = np.mean(perdidas)

# Mostrar resultados
print(f"Último valor de depósitos: Bs. {ultimo_valor:,.2f}")
print(f"VaR 95% Monte Carlo: Bs. {VaR_95:,.2f}")
print(f"VaR 98% Monte Carlo: Bs. {VaR_98:,.2f}")
print(f"VaR 99% Monte Carlo: Bs. {VaR_99:,.2f}")
print(f"Media de Pérdidas Simuladas: Bs. {media_perdida:,.2f}")

# Gráfico
plt.figure(figsize=(12, 6))
sns.histplot(perdidas, bins=50, kde=True, color='mediumslateblue')

# Líneas de VaR y media
plt.axvline(VaR_95, color='green', linestyle='--', linewidth=2, label=f'VaR 95%: Bs. {VaR_95:,.2f}')
plt.axvline(VaR_98, color='gold', linestyle='--', linewidth=2, label=f'VaR 98%: Bs. {VaR_98:,.2f}')
plt.axvline(VaR_99, color='red', linestyle='--', linewidth=2, label=f'VaR 99%: Bs. {VaR_99:,.2f}')
plt.axvline(media_perdida, color='black', linestyle='-.', linewidth=2, label=f'Media pérdida: Bs. {media_perdida:,.2f}')

# Título y etiquetas
plt.title('Distribución de Pérdidas Simuladas - Método Monte Carlo')
plt.xlabel('Pérdida Simulada (Bs.)')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Delimitar claramente el apetito de riesgo, la tolerancia al riesgo y la capacidad de riesgo es fundamental en la gestión de riesgos porque cada uno cumple un rol específico y complementario que permite a la organización manejar de forma efectiva los riesgos inherentes a sus actividades y objetivos.

El apetito de riesgo define el nivel global de riesgo que las Instituciones o Empresas estan dispuestas a aceptar para alcanzar sus metas estratégicas. Es una declaración de la actitud y disposición de la alta dirección y el consejo respecto a cuánto riesgo consideran razonable asumir en la búsqueda de crecimiento, innovación o valor. Este nivel sirve como guía para la toma de decisiones estratégicas y para alinear la cultura organizacional con los objetivos de negocio.

La tolerancia al riesgo, en cambio, es un concepto más específico y cuantificable que establece los límites aceptables de variación en relación con objetivos particulares. Representa el rango dentro del cual las Instituciones o Empresas pueden operar sin que se consideren necesarias acciones correctivas inmediatas. Definir la tolerancia permite monitorear y controlar el desempeño frente a riesgos concretos, estableciendo umbrales que alertan cuando el riesgo se acerca o supera niveles críticos.

Finalmente, la capacidad de riesgo es el máximo nivel de riesgo que las Instituciones o Empresas pueden soportar sin comprometer su viabilidad o sostenibilidad financiera. Este límite está determinado por factores internos como los recursos, capital, estructura financiera y entorno externo. Conocer la capacidad es vital para evitar sobreexposición y garantizar que la asunción de riesgos se mantenga dentro de niveles manejables.

En conjunto, delimitar estos tres conceptos permite a las Instituciones Financieras principalmente establecer un marco claro para la gestión del riesgo, facilitando la asignación eficiente de recursos, la definición de políticas y controles adecuados, y la toma de decisiones informadas que equilibran la búsqueda de oportunidades con la protección frente a pérdidas inaceptables.

## **Módelos Univariados Garch GJR Garch**


El modelo GARCH (Generalized Autoregressive Conditional Heteroskedasticity) describe la varianza condicional de una serie de tiempo como función de valores pasados de la varianza y de los errores al cuadrado, capturando la volatilidad variable en el tiempo y los clusters de volatilidad.

El modelo GJR-GARCH es una extensión del GARCH univariado que incorpora el efecto de apalancamiento o asimetría, permitiendo que los shocks negativos tengan un impacto distinto (generalmente mayor) sobre la volatilidad futura en comparación con los positivos.

El modelo GARCH(1,1) asume que la volatilidad depende de shocks recientes (errores pasados al cuadrado) y de la volatilidad pasada, capturando la característica de agrupamiento de volatilidad (clusters), donde periodos de alta volatilidad tienden a seguirse de más volatilidad.

El modelo GJR-GARCH es una extensión del GARCH que incorpora asimetrías o efectos de apalancamiento, permitiendo que los shocks negativos (caídas) tengan un impacto diferente y generalmente mayor sobre la volatilidad futura que los shocks positivos. Esto refleja mejor la realidad financiera, donde las malas noticias suelen aumentar más la incertidumbre que las buenas.

**Interpretación de Resultados:**

Los resultados muestran que el modelo GJR-GARCH mostró una media negativa significativa en los log-retornos, indicando una tendencia a la baja en el periodo analizado.

El parámetro alpha (impacto inmediato de shocks) fue significativo en GJR-GARCH, mientras que el parámetro asimétrico gamma no lo fue, sugiriendo que en la serie el efecto diferencial de caídas frente a subidas no es estadísticamente claro.

En el modelo GARCH, la volatilidad mostró persistencia moderada y reaccionó a los eventos extremos, con picos de volatilidad que permanecen por varios periodos, capturando la memoria del riesgo.

Los gráficos ilustran cómo la volatilidad condicional estimada por ambos modelos responde a eventos extremos, pero el GJR-GARCH puede capturar mejor la posible asimetría en la reacción del mercado.

En general, el modelo GJR-GARCH es preferible para la gestión de riesgos cuando los datos muestran asimetrías en la volatilidad, es decir, cuando las caídas tienen un efecto mayor en la volatilidad futura que las subidas. Esto es común en mercados financieros, donde las malas noticias generan mayor incertidumbre.

Si los datos no presentan un efecto asimétrico significativo, el modelo GARCH puede ser suficiente, ofreciendo un balance entre simplicidad y capacidad predictiva.

Además, el GJR-GARCH, al modelar colas pesadas mediante distribuciones como la t de Student, mejora la captura de eventos extremos, crucial para estimaciones robustas de riesgo en escenarios adversos.

El modelo GJR-GARCH es más realista porque reconoce que las malas noticias o caídas fuertes suelen aumentar mucho más el riesgo que las buenas noticias. Por eso, sus estimaciones de pérdidas son más altas, especialmente en situaciones extremas.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

# --- Parámetros base ---
ultimo_valor = df['Depositos'].iloc[-1]

# Extraer volatilidad condicional estimada (serie temporal)
garch_volatility = garch_fit.conditional_volatility
gjr_garch_volatility = gjr_garch_fit.conditional_volatility

# Extraer media condicional estimada (constante en este caso)
mu_garch = garch_fit.params['mu']
mu_gjr = gjr_garch_fit.params['mu']

# --- Valores críticos para distribución normal (GARCH) ---
z_95 = norm.ppf(0.05)   # ~ -1.645
z_98 = norm.ppf(0.02)   # ~ -2.054
z_99 = norm.ppf(0.01)   # ~ -2.326

# --- Valores críticos para distribución t (GJR-GARCH) ---
nu = gjr_garch_fit.params['nu']  # grados de libertad estimados
t_95 = t.ppf(0.05, df=nu)
t_98 = t.ppf(0.02, df=nu)
t_99 = t.ppf(0.01, df=nu)

# --- Cálculo del VaR ---

# Para GARCH (normal)
VaR_GARCH_95 = ultimo_valor - ultimo_valor * np.exp(mu_garch + z_95 * garch_volatility)
VaR_GARCH_98 = ultimo_valor - ultimo_valor * np.exp(mu_garch + z_98 * garch_volatility)
VaR_GARCH_99 = ultimo_valor - ultimo_valor * np.exp(mu_garch + z_99 * garch_volatility)

# Para GJR-GARCH (t de Student)
VaR_GJR_GARCH_95 = ultimo_valor - ultimo_valor * np.exp(mu_gjr + t_95 * gjr_garch_volatility)
VaR_GJR_GARCH_98 = ultimo_valor - ultimo_valor * np.exp(mu_gjr + t_98 * gjr_garch_volatility)
VaR_GJR_GARCH_99 = ultimo_valor - ultimo_valor * np.exp(mu_gjr + t_99 * gjr_garch_volatility)

# --- Graficar VaR para GARCH ---
plt.figure(figsize=(12,6))
plt.plot(VaR_GARCH_95, label="VaR GARCH 95%", color='green', linestyle='-')
plt.plot(VaR_GARCH_98, label="VaR GARCH 98%", color='yellow', linestyle='-')
plt.plot(VaR_GARCH_99, label="VaR GARCH 99%", color='red', linestyle='-')
plt.title('Valor en Riesgo (VaR) - Modelo GARCH')
plt.xlabel('Periodo')
plt.ylabel('VaR (Pérdida esperada en Bs.)')
plt.legend()
plt.grid(True)
plt.show()

# --- Graficar VaR para GJR-GARCH ---
plt.figure(figsize=(12,6))
plt.plot(VaR_GJR_GARCH_95, label="VaR GJR-GARCH 95%", color='green', linestyle='-')
plt.plot(VaR_GJR_GARCH_98, label="VaR GJR-GARCH 98%", color='yellow', linestyle='-')
plt.plot(VaR_GJR_GARCH_99, label="VaR GJR-GARCH 99%", color='red', linestyle='-')
plt.title('Valor en Riesgo (VaR) - Modelo GJR-GARCH')
plt.xlabel('Periodo')
plt.ylabel('VaR (Pérdida esperada en Bs.)')
plt.legend()
plt.grid(True)
plt.show()

# --- Mostrar resultados finales ---
print("Últimos valores de VaR GARCH y GJR-GARCH:")
print(f"VaR GARCH 95%: Bs. {VaR_GARCH_95.iloc[-1]:,.2f}")
print(f"VaR GARCH 98%: Bs. {VaR_GARCH_98.iloc[-1]:,.2f}")
print(f"VaR GARCH 99%: Bs. {VaR_GARCH_99.iloc[-1]:,.2f}")

print(f"VaR GJR-GARCH 95%: Bs. {VaR_GJR_GARCH_95.iloc[-1]:,.2f}")
print(f"VaR GJR-GARCH 98%: Bs. {VaR_GJR_GARCH_98.iloc[-1]:,.2f}")
print(f"VaR GJR-GARCH 99%: Bs. {VaR_GJR_GARCH_99.iloc[-1]:,.2f}")

# --- Diferencias entre modelos ---
diferencias_95 = VaR_GARCH_95 - VaR_GJR_GARCH_95
diferencias_98 = VaR_GARCH_98 - VaR_GJR_GARCH_98
diferencias_99 = VaR_GARCH_99 - VaR_GJR_GARCH_99

print("\nDiferencias entre VaR GARCH y GJR-GARCH para los diferentes niveles de confianza:")
print(f"Diferencia VaR 95%: Bs. {diferencias_95.iloc[-1]:,.2f}")
print(f"Diferencia VaR 98%: Bs. {diferencias_98.iloc[-1]:,.2f}")
print(f"Diferencia VaR 99%: Bs. {diferencias_99.iloc[-1]:,.2f}")

```

Tras un exhaustivo análisis de diferentes metodologías y modelos para el cálculo del Valor en Riesgo (VaR), se ha logrado no solo comprender la importancia estadística de esta medida, sino también su relevancia práctica dentro del marco de la gestión integral de riesgos. La elección del modelo más adecuado depende en gran medida del perfil de riesgo institucional, la calidad y naturaleza de los datos disponibles, así como de los recursos tecnológicos y el enfoque estratégico que adopte la organización.

En este sentido, la simulación de Monte Carlo destaca por su flexibilidad y capacidad para representar escenarios complejos y no lineales, facilitando una visualización clara y comprensible del riesgo al que se expone la institución. Su uso es especialmente valioso cuando se requiere modelar distribuciones de pérdidas que no se ajustan a supuestos paramétricos estrictos y cuando se busca una aproximación robusta a la incertidumbre inherente.

No obstante, en presencia de series temporales financieras que exhiben cambios bruscos y persistentes en la volatilidad, el modelo GJR-GARCH se posiciona como una herramienta superior para la estimación del VaR. Su capacidad para capturar el efecto asimétrico -es decir, la mayor repercusión de eventos negativos extremos sobre la volatilidad futura- permite una medición más realista y conservadora del riesgo. Esto es fundamental para evitar la subestimación del riesgo de liquidez o de mercado, especialmente en instituciones donde la exposición a caídas abruptas puede comprometer la estabilidad financiera.

Por tanto, desde una perspectiva profesional y práctica, la recomendación es adoptar un enfoque híbrido o complementario: utilizar la simulación de Monte Carlo para escenarios generales y análisis exploratorios, y aplicar modelos GJR-GARCH para el monitoreo continuo y la gestión del riesgo en entornos con alta volatilidad y eventos extremos. Esta combinación permite equilibrar la precisión estadística con la aplicabilidad operativa, fortaleciendo la capacidad de la institución para anticipar, medir y mitigar riesgos de manera efectiva.

Finalmente, es imprescindible que la selección y aplicación de cualquier modelo de VaR se acompañe de un riguroso proceso de validación, pruebas de estrés y revisión periódica, asegurando que las estimaciones se mantengan alineadas con la evolución del mercado y las condiciones internas de la organización.
